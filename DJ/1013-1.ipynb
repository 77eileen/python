```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 예시 데이터 (원래 df로 바꿔서 사용하세요)
df = pd.DataFrame({
    'weight': [3500, 2500, 3000, 4000],
    'horsepower': [130, 90, 110, 150]
})

# 표준화
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['weight', 'horsepower']])
df_scaled = pd.DataFrame(scaled, columns=['weight', 'horsepower'])

# ✅ 평균 비교
mean_compare = pd.DataFrame({
    '원본 평균': df.mean(),
    '표준화 후 평균': df_scaled.mean()
})

# ✅ 표준편차 비교
std_compare = pd.DataFrame({
    '원본 표준편차': df.std(),
    '표준화 후 표준편차': df_scaled.std()
})

# ✅ 그래프 그리기 (보기 좋게 설정)
plt.figure(figsize=(8,4))
mean_compare.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('표준화 전후 평균 비교')
plt.ylabel('평균 값')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
std_compare.plot(kind='bar', color=['lightgreen', 'orange'])
plt.title('표준화 전후 표준편차 비교')
plt.ylabel('표준편차 값')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

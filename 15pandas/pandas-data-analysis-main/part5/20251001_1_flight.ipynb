```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info()
# dtypes: category(1), int64(2)
#  범주형 데이터(categorical data): pandas에서 제공하는 특수 자료형 / 값이 한정된 몇 개의 카테고리(범주) 중 하나에 속하는 데이터
```

```python
flights.T
```

```python
flights.pivot(index='year', columns='month', values='passengers')
```

```python
# 연도별 승객수의 변화 lineplot - sns
import matplotlib.pyplot as plt
import seaborn as sns
sns.lineplot(data=flights, x='year', y='passengers', errorbar=('ci', 95))  
                                                 #▲ errorbar=('ci', 95) 95% 신뢰구간을 가짐 /신뢰구간을 없애고 싶으면, errorbar=None
# sns 시각화는 신뢰구간도 같이 표시하는 경우
    #데이터가 변동될 수 있는 구간
#신뢰구간 폭 = 데이터 분산 / 샘플 수에 의해 결정돼요.
#신뢰구간이 넓으면 불확실성이 크다 ==> 데이터 분산이 크거나 샘플수가 작을 경우
#신뢰구간이 좁으면 안정적 ==> 데이터가 모여있거나 샘플수가 많을경우

plt.show()
```

```python
# 연도별 월별 승객수의 변화 시각화: lineplot - sns
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,5))
sns.lineplot(data=flights, x='year', y='passengers', hue='month', errorbar=None)  
plt.legend(loc='best')
plt.show()f
```

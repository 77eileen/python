```python
import pandas as pd

# pd.date_range()는
# start + end + freq 조합,
# start + periods + freq 조합,
# start + end + periods 조합
# 이렇게 3개 중 하나만 정상적으로 쓸 수 있어요.


# Timestamp의 배열 만들기 - 월 간격, 월의 시작일 기준
ts_ms = pd.date_range(start='2025-01-01',    # 날짜 범위의 시작
                   end='2025-10-02',                 # 날짜 범위의 끝
                   periods=100,                # 생성할 Timestamp의 개수
                #    freq='MS',                # 시간 간격 (MS: 월의 시작일)
                   tz='Asia/Seoul')          # 시간대(timezone)

# print(ts_ms)
# print(ts_ms.strftime('%y-%m-%d'))
print(type(ts_ms))  #<class 'pandas.core.indexes.datetimes.DatetimeIndex'> : ts_ms ==> 문자열

data = {'date': ts_ms.strftime('%Y-%m-%d')}
df = pd.DataFrame(data)
df.date = pd.to_datetime(df.date)
print(type(df))
df
```

```python
#연도/월/요일을 추출해서 새로운 컬럼으로 넣기
df['year'] = df.date.dt.year
df['month'] = df.date.dt.month
df['day_int'] = df.date.dt.weekday  #0:월요일

#day_int -> day_str 함수 만들기
def convert_week(x):
    temp = ['월', '화', '수', '목', '금', '토', '일']
    return temp[x]
df['day_str'] = df['day_int'].apply(lambda x: convert_week(x))
df.head()


# #day_weekend : 주말 1, 나머지는 0
# def convert_weekend(x):
#     if x == 5 or x==6:
#         return 1
#     else:
#         return 0

# df['day_weekend'] = df['day_int'].apply(lambda x: convert_weekend(x))

df['day_weekend'] = df['day_int'].apply(lambda x: 1 if x in ['토', '일'] else 0)

df.head()

```

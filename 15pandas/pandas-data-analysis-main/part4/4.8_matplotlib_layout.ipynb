```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# 예제 2.7의 재무제표 API를 활용하여 수집한 데이터
samsung_revenue = pd.read_csv("./data/삼성전자_분기별_매출액.csv")
samsung_revenue = samsung_revenue.sort_values('quarter')
samsung_revenue
```

```python
# 비어 있는 figure 객체 생성 (Axes 객체 없음)
fig = plt.figure()
print(fig)
```

```python
# figure 객체 1개, Axes 객체 1개 
fig, axe = plt.subplots()   #subplots() 화면분할 함수!!
print(fig)
print(axe)
```

```python
# axe 객체에 차트 추가
fig, axe = plt.subplots(figsize=(8, 2))
axe.plot(samsung_revenue['quarter'], samsung_revenue['value'])
plt.show()
```

```python
# figure 객체 1개, Axes 객체 2개 (1x2 grid)
fig, axes = plt.subplots(1, 2)  #subplots() 화면분할 함수!! (1행,2열)로 함수그래프 보여줌
print(fig)
print(axes)
```

```python
# axe 객체에 차트 추가
fig, axes = plt.subplots(1, 2, figsize=(12, 2))
axes[0].plot(samsung_revenue['quarter'], samsung_revenue['value'])
samsung_revenue['value'].plot(ax=axes[1])
plt.show()
```

```python
# figure 객체 1개, Axes 객체 4개 (2x2 grid)
fig, axes = plt.subplots(2, 2)
print(fig)
print(axes)
```

```python
# axe 객체에 차트 추가
fig, axes = plt.subplots(2, 2, figsize=(12, 2), constrained_layout=True)
axes[0, 1].plot(samsung_revenue['quarter'], samsung_revenue['value'])
axes[1, 0].plot(samsung_revenue['quarter'], samsung_revenue['value'])
plt.show()
```

```python
# figure 객체 1개, Axes 객체 3개 (왼쪽 2개, 오른쪽 1개)
fig, axes = plt.subplot_mosaic([['top_left', 'right'],
                               ['bottom_left', 'right']])
print(fig)
print(axes)
```

```python
# axe 객체에 차트 추가
fig, axes = plt.subplot_mosaic([['top_left', 'right'],
                               ['bottom_left', 'right']],
                               figsize=(12, 4))
axes['right'].plot(samsung_revenue['quarter'], samsung_revenue['value'])
plt.show()
```

```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# .data 데이터를 확인
from glob import glob
glob('./data/*.*')
glob('./data/*.xlsx')
## ['./data\\경기도인구데이터.xlsx',
#  './data\\남북한발전전력량.xlsx',
#  './data\\서울지역_대학교_위치.xlsx',
#  './data\\시도별_전출입_인구수.xlsx']
glob('./data/*.xlsx')[0]
df2 = pd.read_excel(glob('./data/*.xlsx')[0])

#시군 나누기

# si,gu = "수원시장안구".split('시')
# si,gu
# si+'시'
# df2['구분'].map(lambda x: x.split('시'))
# df2['시'] = df2['구분'].apply(lambda x: x.split('시')[0]+'시')
# df2['군'] = df2['구분'].apply(lambda x: x.split('시')[1] if len(x.split('시')) >=2 else x)
# df2

# '시흥시'.find('시')  #==> index : 0
# '시흥시'[0+1:].find('시')  #==>  index : 1

# findname = '시흥시'
# lastindex=0
# for idx, i in enumerate(findname):
#     if i =='시':
#         lastindex = idx

# findname[:lastindex+1], findname[lastindex+1:]

def split_si(x):
    lastindex = None
    for idx, i in enumerate(x):
        if i =='시':
            lastindex = idx
    if lastindex:
        return x[:lastindex+1]
    else:
        return x

def split_gu(x):
    lastindex = None
    for idx, i in enumerate(x):
        if i =='시':
            lastindex = idx
    if lastindex:
        return x[lastindex+1:]
    else:
        return x

df2['시'] = df2['구분'].apply(split_si)
df2['구'] =df2['구분'].apply(split_gu)
df2
df3=df2.drop(columns=['구분']) #구분제외
df3
```

```python
# 시별 통계
df4 = df3.groupby('시').sum().iloc[:, :-1]  #군을 제외


# matplotlib 한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_path = "./data/malgun.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# subplots (subplot 혼동주의!!!!!!)
fig,axes = plt.subplots(2,2, figsize=(20,10))
axes[0,0].plot(df4.columns, df4.iloc[0].values) #axes[0,0]  제1사분면 / 가평군데이터
axes[0,0].set_title('가평군')
axes[0,1].plot(df4.columns, df4.iloc[1].values) #고양시
axes[0,1].set_title('고양시')
axes[1,0].plot(df4.columns, df4.iloc[2].values) #과천시
axes[1,0].set_title('과천시')
axes[1,1].plot(df4.columns, df4.iloc[3].values) #광명시
axes[1,1].set_title('광명시')
plt.tight_layout()  # 겹치지 않게 정리
plt.show()
# df4
```

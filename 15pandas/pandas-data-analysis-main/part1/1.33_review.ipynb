```python
import pandas as pd
import seaborn as sns
```

```python
mpg=sns.load_dataset('mpg')
mpg.head()
```

```python
# groupby : cylinders별 마력의 평균
    # name별 연식의 평균
    # name이 공백을 기준으로 제조사[0] - 모델[나머지]
    # 제조사 컬럼을 만들고 제조사를 추가 
# 필터 : mpg가 15이상인 모델 
# index를 재구성 : origin
# iloc, loc 를 이용해서 데이터를 row, column 다양하게 출력  
```

```python
# groupby : cylinders별 마력의 평균
mpg.groupby('cylinders', observed=True)['horsepower'].mean()
```

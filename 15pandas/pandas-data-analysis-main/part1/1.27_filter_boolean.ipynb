```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# # titanic 데이터셋 로딩
titanic = sns.load_dataset('titanic')            #titanic : seaborn 자체에 탑재되어 있음.


# # 행 인덱스 0~9 범위에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
# df = titanic.loc[0:9, ['age','fare']]

# 데이터프레임 출력
titanic.head()

# drop : 중복되는 정보 / Nan이 많이 기재된 컬럼을 삭제
titanic_simple = titanic.drop(columns=['alive', 'class', 'who', 'embark_town', 'deck']) #(columns=[~~], inplace=True) inplace=True이거쓰면 원본에 수정된 사항 반영!
titanic_simple.head()
```

```python
# 조건식 적용 (20세 미만: age < 20)
titanic['age'] < 20   #시리즈 형태로 출력됨. (인덱스와 각 인덱스에 조건 만족 여부에 따라 T F 로 표기)
titanic[titanic['age'] < 20]
```

```python
# 불린 인덱싱 (20세 미만의 조건을 충족하는 행을 필터링)
titanic_under20 = titanic[titanic['age'] < 20]   #164명
titanic_under20
# 20세 미만 승객중 생존한 승객만 필터링 : 79명
titanic_under20_survived=titanic_under20[titanic_under20['survived']==1]
titanic_under20_survived=titanic_under20_survived.rename(columns={'sex':'gender'})  #gender로 칼럼이름 변경
# 20세 미만 승객중 생존한 승객 중 여성승객 : 53명
titanic_under20_survived_female=titanic_under20_survived[titanic_under20_survived['gender'] == 'female']
titanic_under20_survived_female.head(3)
# len(titanic_under20_survived_female)

# age < 20 그리고 gender == female
# (titanic['age'] < 20) and (titanic['sex'] == 'female') 
# ==> 오류 발생 (The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all().)
# ==> and or 를 사용할 때는 T,F중 하나값만 가져야 하는데, 지금은 True, False 여러개가 있으므로(시리즈로.. 각행마다 F,T다르므로).. (age<20 해당값, female 해당값도 여러개라서..?)
 
mask = (titanic['age'] < 20) & (titanic['sex'] == 'female')   # and는 안되고 & 로 가능함.
mask
titanic[mask]
```

```python
# 불린 인덱싱 (loc 인덱서 활용)
print(titanic.loc[titanic['age'] < 20].head(3))
print(titanic[titanic['age'] < 20].head(3))
```

```python
# 불린 인덱싱 (NOT 논리연산 : ~ ) 
titanic.loc[~(titanic['age'] < 20)].head(3) #20살 미만이 아닌 == 20살 이상
```

```python
# 나이가 10대(10~19세)인 승객 (AND 논리연산)
mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic[mask1]
df_teenage.head()
```

```python
# 나이가 10세 미만(0~9세)이고 여성인 승객 (AND 논리연산, loc 인덱서)
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
df_female_under10 = titanic.loc[mask2]
df_female_under10.head()
```

```python
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객 (OR 논리연산)
# age, sex, alone 열만 선택 (loc 인덱서, fancy indexing)
mask3 = (titanic.age < 10) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3, ['age', 'sex', 'alone']]
df_under10_morethan60.head()
```

```python
# 불린 시리즈를 새로운 열로 추가
titanic['age_under_20'] = titanic['age'] < 20
titanic.head()
```

```python
# len(titanic)

#1등석, 2등석, 3등석 pclass : 각 선실별 승객수
ti_class3=titanic[titanic['pclass'] == 3]
len(ti_class3)
print(  len(titanic[titanic['pclass'] == 3]) )

ti_class2=titanic[titanic['pclass'] == 2]
len(ti_class2)

ti_class1=titanic[titanic['pclass'] == 1]
len(ti_class1)

print(len(ti_class1), len(ti_class2), len(ti_class3))
print(f'1등석 {len(ti_class1)}, 2등석 {len(ti_class2)}, 3등석 {len(ti_class3)}')

```

```python
import pandas as pd

# 데이터프레임 변환
df = pd.read_excel('./data/남북한발전전력량.xlsx')
df
```

```python
# 남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0, 5], 2:]  # 0에 있는 남한 / 5에 있는 북한 / 2: 1990년부터 쭈욱 데이터 가져옴. 

# 행 인덱스 변경
df_ns.index = ['South','North'] 

# 발전량 데이터의 자료형을 문자열에서 정수형으로 변환
df_ns = df_ns.astype(int)                 

# 데이터프레임 출력
df_ns.head()
```

```python
# 자료형 확인
df.info()    #==> object
df_ns.info() #==> int
```
